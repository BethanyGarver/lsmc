```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm as LN
import pynbody as pb
from dlnpyutils import coords
```

```python
xmin = None
xmax = None
ymin = None
ymax = None
zmin = None
zmax = None
for i in range(61):
    sim = pb.load('run1back6000gas.{:04}0'.format(i))
    mws = sim.s[:240000]
    sim['pos'] = sim['pos'] - np.array([np.mean(mws['x']),np.mean(mws['y']),np.mean(mws['z'])])
    if xmin is None or np.min(sim.s['x'])<xmin:
        xmin = np.min(sim.s['x'])
    if xmax is None or np.max(sim.s['x'])>xmax:
        xmax = np.max(sim.s['x'])
    if ymin is None or np.min(sim.s['y'])<ymin:
        ymin = np.min(sim.s['y'])
    if ymax is None or np.max(sim.s['y'])>ymax:
        ymax = np.max(sim.s['y'])
    if zmin is None or np.min(sim.s['z'])<zmin:
        zmin = np.min(sim.s['z'])
    if zmax is None or np.max(sim.s['z'])>zmax:
        zmax = np.max(sim.s['z'])

for i in range(61):
    sim = pb.load('run1back6000gas.{:04}0'.format(i))
    mws = sim.s[:240000]
    sim['pos'] = sim['pos'] - np.array([np.mean(mws['x']),np.mean(mws['y']),np.mean(mws['z'])])
    plt.figure(figsize=(12,3.5))
    plt.subplot(1,3,1)
    plt.hist2d(sim.s['x'],sim.s['y'],bins=100,norm=LN(),range=[[xmin,xmax],[ymin,ymax]])
    plt.xlim(xmin,xmax)
    plt.ylim(ymin,ymax)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.subplot(1,3,2)
    plt.hist2d(sim.s['x'],sim.s['z'],bins=100,norm=LN(),range=[[xmin,xmax],[zmin,zmax]])
    plt.xlim(xmin,xmax)
    plt.ylim(zmin,zmax)
    plt.xlabel('x')
    plt.ylabel('z')
    plt.subplot(1,3,3)
    plt.hist2d(sim.s['y'],sim.s['z'],bins=100,norm=LN(),range=[[ymin,ymax],[zmin,zmax]])
    plt.xlim(ymin,ymax)
    plt.ylim(zmin,zmax)
    plt.xlabel('y')
    plt.ylabel('z')
    plt.suptitle('{} Gyr'.format(round(i/10,1)))
```

```python
for i in range(60):
    sim = pb.load('run1back6000gas.{:04}0'.format(i+1))
    mws = sim.s[:240000]
    sim['pos'] = sim['pos'] - np.array([np.mean(mws['x']),np.mean(mws['y']),np.mean(mws['z'])])
    lsmc = sim.s[240000:]
    lsmc['l'],lsmc['b'],lsmc['d'] = coords.xyz2lbd(lsmc['x'],lsmc['y'],lsmc['z'])
    lsmc['mlon'],lsmc['mlat'] = coords.gal2mag(lsmc['l'],lsmc['b'])
    lsmc['mlon'][(lsmc['mlon']>180)] = lsmc['mlon'][(lsmc['mlon']>180)]-360
    plt.figure()
    plt.hist2d(lsmc['mlon'],lsmc['mlat'],bins=100,norm=LN(),range=[[-180,180],[-90,90]])
    plt.plot([0,0],[-90,90],'k-')
    plt.plot([-180,180],[0,0],'k-')
    plt.xlim(180,-180)
    plt.ylim(-90,90)
    plt.xlabel('MS Longitude')
    plt.ylabel('MS Latitude')
    plt.title('{} Gyr'.format(round((i+1)/10,1)))
```
